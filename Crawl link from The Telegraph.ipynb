```python
import requests 
from bs4 import BeautifulSoup
```


# Crawl link and dump into file.json

```python
def crawl_fn(url, lst):
    response = requests.get(url)   #Get html
    soup = BeautifulSoup(response.text, "html.parser") 
    grid = soup.find("ul", class_ = "article-list__list grid")  #Find part containing links
    link = grid.find_all(attrs={"data-track-wrapper": "article-list "})
    for element in link:
        a = element.find("a")    #Find links
        lst.append("https://www.telegraph.co.uk" + a["href"])
    return lst
```

```python
url_sample = "https://www.telegraph.co.uk/business/economy/page-50/"
```
